```python
import numpy as np
import pandas as pd

import transformers
from datasets import Dataset,load_dataset,load_from_disk, DatasetDict
from transformers import AutoTokenizer, AutoModelForSequenceClassification, BertForSequenceClassification, BertTokenizer, Trainer
from scipy.stats import pearsonr
import sklearn.metrics
from datasets import load_metric, load_dataset
from sklearn.metrics import mean_squared_error
```

```python
dataset = load_dataset("PlanTL-GOB-ES/sts-es")
dataset
```

```python
len(dataset["test"]["sentence1"] + list(df_splits[0]["sentence1"]))
```

```python
complete_dataset = {"train":{},
                    "validation":{},
                    "test":{}}

for i, partition in enumerate(dataset):
  complete_dataset[partition]['sentence1'] = dataset[partition]["sentence1"] + list(df_splits[i]["sentence1"])
  complete_dataset[partition]['sentence2'] = dataset[partition]["sentence2"] + list(df_splits[i]["sentence2"])
  complete_dataset[partition]['label'] = dataset[partition]["label"] + list(df_splits[i]["label"])
  #complete_dataset["id"] = [i for i in range(len(dataset[partition]['label']))]
```

```python
ds = DatasetDict()

ds['train'] = Dataset.from_dict(complete_dataset["train"])
ds['validation'] = Dataset.from_dict(complete_dataset["validation"])
ds['test'] = Dataset.from_dict(complete_dataset["test"])
```
